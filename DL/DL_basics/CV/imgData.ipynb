## **Classification**

<img src="../CV/assets/img/Screenshot 2023-08-10 at 2.27.04 PM.png" style="width: 800px; height: ;">

## **single object**

- **Classification (분류):**
분류는 컴퓨터 비전 분야에서 주어진 입력 이미지를 사전에 정의된 클래스 레이블 중 하나로 분류하는 작업

- **Localization:** 
특정한 객체가 하나의 이미지안에서 어디에 있는지 Bounding Box로 지정해주는 작업

## **Multiple objects**

<img src="../CV/assets/img/Screenshot 2023-08-10 at 2.35.58 PM.png" style="width: 800px; height: ;">

#### **Object Detection(BBOX + Classification)**

- 이미지 내, 객체를 검출  

- 각 객체의 위치와 BBox를 표시  

- 각 객체의 클래스 레이블을 붙히는 작업

## **Segmentation**

<img src="../CV/assets/img/Screenshot 2023-08-10 at 2.42.14 PM.png" style="width: 800px; height: ;">

- 픽셀 단위에서 클래스를 찾는 태스크

- 이미지를 작은 영역으로 나누고 각 영역을 객체 또는 배경으로 분류

- Semantic Segmentation은 픽셀을 객체의 클래스로 분류하여 이미지를 분할  

- Instance Segmentation은 개별 객체들을 개별적으로 식별하여 분할, 객체 간의 겹침과 부분적 가려짐 문제를 처리함으로서 보다 정확한 객체 분할을 제공
