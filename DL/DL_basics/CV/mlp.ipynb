# **Multi-Layer Perceptron**

### **딥러닝이란?**

<img src="../CV/assets/img/Screenshot 2023-08-10 at 4.06.49 PM.png" style="width: 800px; height: ;">

*입력층, 은닉층, 출력층 등 다양한 층으로 구성된 인공신경망*

> **은닉층 (Hidden Layers):** 딥러닝에서 '딥'이란 은닉층의 개수를 의미하고, 은닉층은 데이터의 특징을 추상화하고 변환하는 역할을 수행

### **퍼셉트론의 구조 및 역할**

<img src="../CV/assets/img/Screenshot 2023-08-10 at 4.27.51 PM.png" style="width: 800px; height: ;">


- **입력 데이터 수신:** 뉴런은 입력층에서 들어오는 데이터를 받아들임  

- **가중치와 편향 적용:** 입력 데이터는 뉴런에 대한 가중치와 편향과 함께 곱해지고 합산

- **활성화 함수 적용:** 가중치와 편향을 적용한 합산 결과는 활성화 함수를 통과하게 됩니다. 활성화 함수는 비선형성을 추가하며, 뉴런의 출력 값을 조정

### **멀티 퍼셉트론의 효과**  

<img src="../CV/assets/img/Screenshot 2023-08-16 at 10.35.49 PM.png" style="width: 800px; height: ;">

- Hidden Layer가 깊어지면 깊어질수록 못찾았던 패턴들을 더 다양하게 찾을 수 있음

### **Representation Learning이란?**

<img src="../CV/assets/img/Screenshot 2023-08-16 at 10.40.54 PM.png" style="width: 800px; height: ;">

<img src="../CV/assets/img/Screenshot 2023-08-10 at 4.40.15 PM.png" style="width: 800px; height: ;">


*"**Representation learning**은 데이터의 특성이나 구조를 자동으로 추출하고 학습하는 과정을 의미함. 이를 통해 원시 데이터를 더 의미 있는 형태로 변환하고, 데이터의 핵심적인 특징을 잘 나타내는 표현을 학습과정"*
