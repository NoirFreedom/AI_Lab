#### **데이터프레임 복사**
##### 참고 | 데이터프레임 값을 변수에 할당하고 변경하려할 경우 에러 발생 -> .copy() 사용

```python
data = df.copy()
data
```

##### **데이터 삭제1 | .drop( 인덱스(행)명 또는 컬럼명, axis=0 or 1)**

```python
data = data.drop(5, axis=0)
data
```

##### **데이터 삭제2 | data.drop(columns=['삭제할 컬럼 명', '여러개 가능'])**

```python
data = df.copy()
data = data.drop(columns=['Unnamed: 0'])
data
```

```python
df = data.copy()
df
```

##### **변수 할당 없이, 변경된 값 적용하기 | inplace=True**
